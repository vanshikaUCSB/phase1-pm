```python
#Answer to Q1 
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display


# Load the CSV file into a DataFrame
df = pd.read_csv('./data/Construction_Data_PM_Tasks_All_Projects.csv') 

# Display the first few rows to understand its structure
df.head()

# Count the number of overdue tasks
overdue_tasks_count = df[df['OverDue'] == True].shape[0]
print("Answer 1: The total number of tasks that are overdue = " , overdue_tasks_count)

# Answer to Q2

# Group by 'Task Group' and 'Status', then count the tasks for each combination
task_group_status_counts = df.groupby(['Task Group', 'Status']).size().unstack().fillna(0)

# Filter the columns to only include 'Open' and 'Closed' statuses
task_group_status_counts = task_group_status_counts[['Open', 'Closed']]
print("Answer 2:")
display(task_group_status_counts)

#Answer  to Q3
colors = {'Open': 'red', 'Closed': 'blue'}

# Plotting the bar chart without stacking for better visibility
ax = task_group_status_counts[['Open', 'Closed']].plot(kind='bar', figsize=(12, 7), color=[colors[col] for col in ['Open', 'Closed']])

# Setting the title and labels
plt.title('Number of Open and Closed Tasks by Task Group')
plt.xlabel('Task Group')
plt.ylabel('Number of Tasks')
plt.legend(title='Status')

# Annotating the bars with the exact numbers
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{int(height)}', (x + width/2, y + height + 10), ha='center', color='black')
print("Answer 3: ")
# Display the plot
plt.tight_layout()
plt.show()
 




 
 #Answer 4

 # Filtering the dataframe to only include overdue tasks
overdue_df = df[df['OverDue'] == True]

# Grouping by 'project' and counting the number of overdue tasks for each project
overdue_counts_by_project = overdue_df.groupby('project').size()

# Plotting the bar chart
ax = overdue_counts_by_project.plot(kind='bar', figsize=(12, 7), color='skyblue')

# Setting the title and labels
plt.title('Number of Overdue Tasks by Project')
plt.xlabel('Project')
plt.ylabel('Number of Overdue Tasks')

# Annotating the bars with the exact numbers
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{int(height)}', (x + width/2, y + height + 5), ha='center', color='black')
print("Answer 4: ")
# Display the plot
plt.tight_layout()
plt.show()
```
